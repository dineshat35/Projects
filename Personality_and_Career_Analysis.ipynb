On the next page the following values were collected.

race	Chosen from a drop down menu. 1=Mixed Race, 2=Arctic (Siberian, Eskimo), 3=Caucasian (European), 4=Caucasian (Indian), 5=Caucasian (Middle East), 6=Caucasian (North African, Other), 7=Indigenous Australian, 8=Native American, 9=North East Asian (Mongol, Tibetan, Korean Japanese, etc), 10=Pacific (Polynesian, Micronesian, etc), 11=South East Asian (Chinese, Thai, Malay, Filipino, etc), 12=West African, Bushmen, Ethiopian, 13=Other (0=missed)
age	entered as text (individuals reporting age < 13 were not recorded)
engnat	Response to "is English your native language?". 1=yes, 2=no (0=missed)
gender	Chosen from a drop down menu. 1=Male, 2=Female, 3=Other (0=missed)
hand	"What hand do you use to write with?". 1=Right, 2=Left, 3=Both (0=missed)

On this page users were also asked to confirm that their answers were accurate and could be used for research. Participants who did not were not recorded).

Some values were calculated from technical information.

country	The participant's technical location. ISO country code.
source	How the participant came to the test. Based on HTTP Referer. 1=from another page on the test website, 2=from google, 3=from facebook, 4=from any url with ".edu" in its domain name (e.g. xxx.edu, xxx.edu.au), 6=other source, or HTTP Referer not provided.

```python
# Create a DataFrame from the data
df = big5.drop('country', axis=1)

# Define a function to calculate the personality trait scores
def calculate_personality_scores(df):
    extroversion = df[['EXT1', 'EXT2', 'EXT3']].mean(axis=1)
    emotional_stability = df[['EST1', 'EST2']].mean(axis=1)
    agreeableness = df[['AGR1', 'AGR2']].mean(axis=1)
    conscientiousness = df[['CSN1']].mean(axis=1)
    openness_to_experience = df[['OPN1']].mean(axis=1)
    return extroversion, emotional_stability, agreeableness, conscientiousness, openness_to_experience

# Calculate the personality trait scores
df['Extroversion'], df['Emotional Stability'], df['Agreeableness'], df['Conscientiousness'], df['Openness to Experience'] = calculate_personality_scores(df)

# Define the ideal job preferences for each personality trait
ideal_job_preferences = {
    'Extroversion': ['Sales', 'Marketing', 'Public Relations'],
    'Emotional Stability': ['Counselor', 'Therapist', 'Social Worker'],
    'Agreeableness': ['Human Resources', 'Customer Service', 'Teaching'],
    'Conscientiousness': ['Project Management', 'Accounting', 'Data Analysis'],
    'Openness to Experience': ['Creative Arts', 'Research', 'Innovation'],
}

# Match individuals to their ideal jobs based on personality trait scores
def match_ideal_jobs(personality_scores):
    ideal_jobs = []
    for trait, score in personality_scores.items():
        if trait in ideal_job_preferences:
            for job in ideal_job_preferences[trait]:
                if score >= 3.5:  # You can adjust the threshold as needed
                    ideal_jobs.append((job, trait))
    return ideal_jobs

# Apply the matching function to each individual in the DataFrame
df['Ideal Jobs'] = df.apply(match_ideal_jobs, axis=1)

# Print the results
print(df[['Ideal Jobs']])
```
