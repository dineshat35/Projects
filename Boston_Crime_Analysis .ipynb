# Part 1

```python
import folium
import requests
import pandas as pd
import numpy as np
from shapely.geometry import Point

# The data I decided to get was the crime data set for boston in the year 2022
crime = pd.read_csv("https://data.boston.gov/dataset/6220d948-eae2-4e4b-8723-2dc8e67722a3/resource/313e56df-6d77-49d2-9c49-ee411f10cf58/download/tmp9b74ohf1.csv", low_memory = False)
# Dropped all rows that have latitude and longitude as either 0 or NaN
na = crime[(crime.Lat == 0) | (crime.Long == 0) | pd.isnull(crime.Lat)| pd.isnull(crime.Long)]
crime.drop(na.index, axis = 0, inplace = True)
crime = crime.reset_index()
crime
```

# Part 2

```python
import folium
from folium.map import *
map_osm = folium.Map((42.3601, -71.0589), zoom_start = 11)
# Map centered around Boston
map_osm
```

# Part 3
The exact data I wanted to show with this map was the breakdown of the number of gun related crime occurrences that were reported on a specific day of the week. I had expected the map to have more markers indicating that a larger number of gun related crimes happened on Friday and Saturday since that is when more people would be out of the safety of their homes. I was incorrect since most of the crimes that ivolved guns firing reported in October happened on Mondays and Sundays. The idea is if you are visiting Boston or you live there it is best not to go out on a Monday or a Sunday because you have a higher chance of being entangled in a gun related crime. 

This map marks the exact longitudinal and latitudinal location of where the gun related crime was reported. The map can be used to find the days that had the least reported crimes within the month of October in Boston.

```python
# Count Occurences of crime that occured on a specific day in the month of October.
in1 = 0
in2 = 0
in3 = 0
in4 = 0
in5 = 0
in6 = 0
in7 = 0

# Get only crimes that had reported gun shots and occured in the 10th month of the year (OCTOBER)
# Make a marker for each day as a different color to see the relationship between gun related crimes
# and days of the week.
for index, row in crime[(crime.MONTH == 10) & (crime.SHOOTING == 1)].iterrows():
    if row.DAY_OF_WEEK == "Monday":
        in1 = in1 + 1
        folium.Marker(location=[row["Lat"], row["Long"]], icon=folium.Icon(color = 'black')).add_to(map_osm)
    if row.DAY_OF_WEEK == "Tuesday":
        in2 = in2 + 1
        folium.Marker(location=[row["Lat"], row["Long"]], icon=folium.Icon(color = 'pink')).add_to(map_osm)
    if row.DAY_OF_WEEK == "Wednesday":
        in3 = in3 + 1
        folium.Marker(location=[row["Lat"], row["Long"]], icon=folium.Icon(color = 'orange')).add_to(map_osm)
    if row.DAY_OF_WEEK == "Thursday":
        in4 = in4 + 1
        folium.Marker(location=[row["Lat"], row["Long"]], icon=folium.Icon(color = 'green')).add_to(map_osm)
    if row.DAY_OF_WEEK == "Friday":
        in5 = in5 + 1
        folium.Marker(location=[row["Lat"], row["Long"]], icon=folium.Icon(color = 'purple')).add_to(map_osm)
    if row.DAY_OF_WEEK == "Saturday":
        in6 = in6 + 1
        folium.Marker(location=[row["Lat"], row["Long"]], icon=folium.Icon(color = 'blue')).add_to(map_osm)
    if row.DAY_OF_WEEK == "Sunday":
        in7 = in7 + 1
        folium.Marker(location=[row["Lat"], row["Long"]], icon=folium.Icon(color = 'red')).add_to(map_osm)
print("Monday Crime Count: " + str(in1))
print("Tuesday Crime Count: " + str(in2))
print("Wednesday Crime Count: " + str(in3))
print("Thursday Crime Count: " + str(in4))
print("Friday Crime Count: " + str(in5))
print("Saturday Crime Count: " + str(in6))
print("Sunday Crime Count: " + str(in7))

map_osm 
```
